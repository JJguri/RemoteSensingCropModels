```python
#Open apsim outputs
con = sqlite3.connect(r'C:\Users\jjojeda\Google Drive\COALAR\Sorgo\0. Calibration\Sorghum_V6.db')
apsim = pd.read_sql("Select * from InitialReport", con)

#Read the Simulations table that has SimulationID matched to SimulationName
Simulations = pd.read_sql("Select * from _Simulations",con)
Simulations.set_index('ID',inplace=True)
apsim.loc[:,'SimulationName'] = [Simulations.loc[apsim.loc[x,'SimulationID'],'Name'] for x in apsim.index]
```

```python
#Open apsim outputs
con2 = sqlite3.connect(r'C:\Users\jjojeda\Google Drive\COALAR\Sorgo\0. Calibration\Sorghum_V6_Australia.db')
apsimAu = pd.read_sql("Select * from InitialReport", con2)

#Read the Simulations table that has SimulationID matched to SimulationName
Simulations = pd.read_sql("Select * from _Simulations",con2)
Simulations.set_index('ID',inplace=True)
apsimAu.loc[:,'SimulationName'] = [Simulations.loc[apsimAu.loc[x,'SimulationID'],'Name'] for x in apsimAu.index]
```

```python
#reduce the dataframes
df=apsim.drop(['CheckpointID', 'SimulationID', 'Zone'],axis=1) 

df['climate'] = df['SimulationName'].str[:6]
df['soil'] = df['SimulationName'].str[7:11]
df['cultivar'] = df['SimulationName'].str[12:15]
df['harvest'] = df['SimulationName'].str[16:18]
df['prevcrop'] = df['SimulationName'].str[19:20]
df['all'] = 'Alldata'
```
